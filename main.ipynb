```python
import pandas as pd
from copy import deepcopy
from catboost import CatBoostRanker, Pool
import numpy as np


pd.set_option('display.max_columns', None)
test_data = pd.read_csv('test_df.csv')
train_data = pd.read_csv('train_df.csv')
```

по анализу могу предположить что фичи 73 74 и 75 не несут никакой смысловой нагрузки, поэтому было принято решение их удалить

```python
train_data = train_data.drop(['feature_73','feature_74','feature_75'], axis=1)
test_data = test_data.drop(['feature_73','feature_74','feature_75'], axis=1)
```

делаю первую итерацию в лоб

```python
X_train = train_data.drop(['search_id'], axis=1).values
y_train = train_data['target'].values
queries_train = train_data['search_id'].values


X_test = test_data.drop(['search_id'], axis=1).values
y_test = test_data['target'].values
queries_test = test_data['search_id'].values

max_relevance = np.max(y_train)
y_train = y_train.astype(np.int64) / max_relevance
y_test = y_test.astype(np.int64) / max_relevance

y_train = y_train.astype(np.int64)
y_test = y_test.astype(np.int64)

train = Pool(
    data=X_train,
    label=y_train,
    group_id=queries_train
)

test = Pool(
    data=X_test,
    label=y_test,
    group_id=queries_test
)

default_parameters = {
    'iterations': 2000,
    'custom_metric': ['NDCG', 'PFound', 'AverageGain:top=10'],
    'verbose': False,
    'random_seed': 0,
}

parameters = {}

def fit_model(loss_function, additional_params=None, train_pool=train, test_pool=test):
    parameters = deepcopy(default_parameters)
    parameters['loss_function'] = loss_function
    parameters['train_dir'] = loss_function
    
    if additional_params is not None:
        parameters.update(additional_params)
        
    model = CatBoostRanker(**parameters)
    model.fit(train_pool, eval_set=test_pool, plot=True)
    
    return model

model = fit_model('RMSE', {'custom_metric': ['NDCG']})

test_data['predictions'] = model.predict(test)
```

заметил, что NDCG сразу стало 1, что является наилучшим параметром. Это странно, решил посмотреть подробнее на тестовую выборку. Ниже смотрю, где в группе есть хотя-бы один клик

```python
counts = test_data[test_data['target'] == 1].groupby('search_id').size()

test_data['have_click'] = test_data['target']==1
(test_data.groupby('search_id')['have_click'].mean() != 0).mean()
```

в 23% searc_id в тестовой выборке есть конверсия

проверим, правильно ли ранжирует модель на самом деле

```python
check = test_data[test_data['search_id'] == 237654][['predictions','target']]
check
```

```python
def realno_li_sami_krutoi(predictions,target):
    ones_number = (target == 1).sum()
    assert ones_number != 0
    best_targets = [t for pred, t in sorted(zip(predictions,target), reverse=True)][:ones_number]
    return sum(best_targets) == ones_number
```

```python
realno_li_sami_krutoi(check['predictions'], check['target'])
```

```python
result = []
for i in counts.reset_index()['search_id']:
    check = test_data[test_data['search_id'] == i][['predictions', 'target']]
    result.append(realno_li_sami_krutoi(check['predictions'], check['target']))
print(result)
```
